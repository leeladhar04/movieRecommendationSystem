```python
ratings=pd.merge(ratings,movies)
```

```python
ratings.head()
```

```python
ratings=ratings.drop(["genres","timestamp"],axis=1)
```

```python
ratings.head()
```

```python
user_ratings=ratings.pivot_table(index=["userId"],columns=["title"],values="rating")
```

```python
user_ratings=user_ratings.dropna(thresh=10,axis=1).fillna(0)
```

```python
user_ratings
```

```python
## similarity_matrix
item__similarity_df=user_ratings.corr(method="pearson")
```

```python
item__similarity_df.head()
```

```python
def get_similar_movies(movie_name):
    scores=item__similarity_df[movie_name]
    scores=scores.sort_values(ascending=False)
    return scores[1:11]
    
```

```python
get_similar_movies("Iron Man (2008)")
```
